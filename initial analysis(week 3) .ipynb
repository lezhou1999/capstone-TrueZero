```python
!pip install openpyxl
import pandas as pd
```

```python
df = pd.read_csv("2024.07.05_CMB.csv")
print(df)
```

```python
massk_value = df['LI1_TankPerc']*8.7
print("Liquid Weight(kilogram) in tank :", massk_value)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df['Time'] = pd.to_datetime(df['TimeStamp'])
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], massk_value, color='blue')
plt.title('Trends in Liquid Weight Over Time')
plt.xlabel('Time')
plt.ylabel('Liquid Weight')
plt.legend()

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))

plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
