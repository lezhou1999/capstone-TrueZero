```python
from CoolProp.CoolProp import PropsSI
import matplotlib.pyplot as plt
```

```python
#import Intial data
V_tank = 32.0  # m^3
Q_dot = 40.7  # Watts
P_max = 1204514.0  # Pa
m_initial = 1680.0  # kg
P_init = 2 * 101325  # Initial pressure in Pa
P_final = 5 * 101325  # Final pressure in Pa
```

```python
# Initial data for plotting
pressures = []
qualities = []
enthalpies = []
entropies = []
internal_energies = []
du_total = 0  # Total change in internal energy
time_total = 0  # Total time
```

```python
quality_initial = 0

# Simulation loop (simplified example)
P_current = P_init
while P_current <= P_final:
    quality = quality_initial + 0.01  # Incremental quality update
    h = PropsSI('H', 'P', P_current, 'Q', quality, 'parahydrogen')
    s = PropsSI('S', 'P', P_current, 'Q', quality, 'parahydrogen')
    u_current = PropsSI('U', 'P', P_current, 'Q', quality, 'parahydrogen')

    # Calculate du and add to total
    if internal_energies:  # Calculate du based on previous u
        du = u_current - internal_energies[-1]
        du_total += du
        dt = du / Q_dot if Q_dot else 0  # Prevent division by zero
        time_total += dt

    # Store data for plotting
    pressures.append(P_current)
    qualities.append(quality)
    enthalpies.append(h)
    entropies.append(s)
    internal_energies.append(u_current)


    print(f"Pressure: {P_current:.2f} Pa,     Quality: {quality:.2f}")
    print(f"Enthalpy: {h:.2f} J/kg,         Entropy: {s:.2f} J/kg/K")
    print(f"Internal Energy: {u:.2f} J/kg")
    print(f"Total Time Duration: {time_total:.2f} seconds")
    print("----------------------------------------------------")


    # Update loop data
    P_current += 50000
    quality_initial += 0.01  # Adjust quality increase as needed

    # Update conditions for the next iteration
    P_current += 50000  # Incremental increase in pressure

    # Output final quality and total time
    print(f"Final Quality: {qualities[-1]:.2f}")
    print(f"Total Time Duration: {time_total:.2f} seconds")
```
